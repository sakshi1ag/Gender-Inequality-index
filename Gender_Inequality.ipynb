# **Let's Know our dataset**
The meaning of the terms used in the dataset are described
below:

---

**Human_development:** Freedom that common people have

**GII:** Gender Inequality Index

**Rank:** Secured in GII

**Maternal_mortality:** Deaths per 100,000 live births

**Adolescent_birth_rate:** Annual number of births to women aged 15-19 years per 1,000 women in that age group

**Seats_parliament:** Share of % seats held by women in the parliament

**F_secondary_education:**Females with at least some secondary education (% ages 25 and older)

**M_secondary_education:** Males with at least some secondary education (% ages 25 and older)

**F_labour_force:**Female - Labour force participation rate (% ages 15 and older)

**M_labour_force:** Male - Labour force participation rate (% ages 15 and older)

#Handling Missing Values

Let's see which method imputes the best.
We will either drop or insert mean, median and mode in the place of missing values and will check the change in the shape of the distribution and change in the variation.


---
As we have seen in the data preprocessing, those have less than 5% values missing, those rows will be dropped and the rest will get some imputed value.


```python
cols=[var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
cols
```

```python
len(df[cols].dropna())/len(df)
```

```python
new_df=df[cols].dropna()
new_df.shape,df.shape
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
df['Seats_parliament'].plot(kind="kde",ax=ax,color='Red')
new_df['Seats_parliament'].plot(kind='kde',ax=ax,color="Blue")
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

```python
a=pd.concat([df['Human_development'].value_counts()/len(df),new_df['Human_development'].value_counts()/len(df)],axis=1)
a.columns=["Original","New"]
a
```

```python
df.dropna(subset=['Human_development','Seats_parliament'],inplace=True)
```

**Filling mean and median values in place of Nan values**

```python
mean=df['Maternal_mortality'].mean()
median=df['Maternal_mortality'].median()
df['mean']=df['Maternal_mortality'].fillna(mean)
df['median']=df['Maternal_mortality'].fillna(median)
```

**Checking the variance of the filled values**

```python
print("The original var is:",df['Maternal_mortality'].var())
print("The mean filled data has var:",df['mean'].var())
print("The median filled data has var:",df['median'].var())
```

**Checking the shape of the distribution**

```python
fig=plt.figure()
ax=fig.add_subplot(111)
df['Maternal_mortality'].plot(kind='kde',ax=ax,color='Red')
df['mean'].plot(kind='kde',ax=ax,color='yellow')
df['median'].plot(kind='kde',ax=ax,color='green')
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

As mean hs less difference in variance as compare to original data values and also the shape of the distribution resembles. Thus, we will choose mean in place of Nan values for the Maternal mortality column.


```python
df['Maternal_mortality']=df['Maternal_mortality'].fillna(value=df['Maternal_mortality'].mean())
df
```

```python
df.drop(['mean','median'],axis=1,inplace=True)
```

```python
df=df.drop(df.index[[3,105]],axis=0)
df
```

```python
df0=pd.DataFrame()
df0['F_secondary_educ1']=df['F_secondary_educ'].fillna(method='ffill')
df0['F_secondary_educ2']=df['F_secondary_educ'].fillna(method='bfill')
df0['F_secondary_educ3']=df['F_secondary_educ'].interpolate()
df0['F_secondary_educ4']=df['F_secondary_educ'].fillna(df['F_secondary_educ'].mean())
df0['F_secondary_educ5']=df['F_secondary_educ'].fillna(df['F_secondary_educ'].median())
fig=plt.figure()
ax=fig.add_subplot(111)
df0['F_secondary_educ1'].plot(kind='kde',ax=ax,color='Red')
df['F_secondary_educ'].plot(kind='kde',ax=ax,color='blue')
df0['F_secondary_educ2'].plot(kind='kde',ax=ax,color='yellow')
df0['F_secondary_educ3'].plot(kind='kde',ax=ax,color='pink')
df0['F_secondary_educ4'].plot(kind='kde',ax=ax,color='green')
df0['F_secondary_educ5'].plot(kind='kde',ax=ax,color='black')
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

# Insights


1.   **Female Secondary Education increases**        
Gender inequality index decreases and thus share in the seats of parliament of women also increases. This is also seen that Adolscent birth rate will go down while there would be no significant difference in the women labour force as women are indulged in some kind of work irrespective of their education levels.

2.   **Male Secondary Education increases**
    
     As Male secondary education increases, the labour force of males have slightly gone down.

3.   **Rank increases**

 If Rank increases, Gender Inequality Index decreases and seats of parliament for women will increase.


4.   The countries **Norway** and **Denmark** are good in low maternal mortality rate and low adolescent birth rate respectively.

5. The country **Canada** has both female and male secondary education high while at the same time country **Rwnda** has the highest seats of parliament for women.
